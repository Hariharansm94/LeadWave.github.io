## 📦 Section 2: Variables — Labeled Storage Boxes

Think of a variable as a **labeled box** in your computer's memory:

```
┌──────────────────┐    ┌─────────────────┐    ┌────────────────────┐
│  name = "Rahul"  │    │   age = 25       │    │  salary = 75000.5  │
```

**4 Main Data Types in Python:**

| Type | Example | What it stores |
|------|---------|----------------|
| `str` | `"Rahul"` | Text / Words |
| `int` | `25` | Whole numbers |
| `float` | `75000.50` | Decimal numbers |
| `bool` | `True` or `False` | Yes / No |

## 📋 Section 3: Lists — Hold Multiple Items in One Variable

A list holds **many values** in a single variable using square brackets `[ ]`:

```
fruits = [ "Apple",  "Mango",  "Banana",  "Grapes" ]
              ↑          ↑          ↑          ↑
           index 0    index 1    index 2    index 3
```

> 🔑 **Key Rule:** Python indexing starts from **0** (not 1)!
> Last item always has index **-1**.


````python
fruits  = ["Apple", "Mango", "Banana", "Grapes", "Orange"]
````

````python
import matplotlib.pyplot as plt
import numpy as np

subjects = ['Math','Physics','Chemistry','Biology','English']
marks = [78,85,67,90,82]

plt.figure(figsize=(8,5))

bars = plt.bar(subjects, marks, color='skyblue', edgecolor='black')

# add values on bars
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y + 1, y,
             ha='center', fontsize=11, fontweight='bold')

# mean line
mean_val = np.mean(marks)
plt.axhline(mean_val, color='red', linestyle='--', linewidth=2,
            label=f'Mean = {mean_val:.1f}')

plt.title("Student Marks by Subject", fontsize=14, fontweight='bold')
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend()

plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

students = ["Aman","Riya","Karan","Neha","Arjun"]
math = [78,85,67,90,82]
science = [80,88,70,86,84]
english = [75,82,65,88,80]

x = np.arange(len(students))
width = 0.25

plt.figure(figsize=(9,5))

# Color-blind friendly palette
c_math = "#0072B2"     # blue
c_science = "#009E73"  # green
c_english = "#E69F00"  # orange

bars1 = plt.bar(x - width, math, width, label="Math", color=c_math)
bars2 = plt.bar(x, science, width, label="Science", color=c_science)
bars3 = plt.bar(x + width, english, width, label="English", color=c_english)

# Value labels on bars
for bars in [bars1, bars2, bars3]:
    for bar in bars:
        h = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, h + 1, f"{h}",
                 ha="center", fontsize=9)

# Labels and styling
plt.xticks(x, students)
plt.xlabel("Students", fontsize=11)
plt.ylabel("Marks", fontsize=11)
plt.title("CLASS COMPARISON", fontsize=7, fontweight="bold")

plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.legend()

# plt.tight_layout()
plt.show()
````
